# Outlier


```python
#Import Library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Reading Dataset dalam format data Excel
df_1 = pd.read_excel('/content/Tungkub_Hourly_PM_New.xlsx')
df_1 = df_1.drop(0)
df_1 = df_1.drop(1)
df_1.head()
```

```python
# Menghitung IQR
Q1 = df_1['Rs'].quantile(0.25)
Q3 = df_1['Rs'].quantile(0.75)
IQR = Q3 - Q1

# Menentukan batas bawah dan batas atas
batas_bawah = Q1 - 1.5 * IQR
batas_atas = Q3 + 1.5 * IQR

# Menandai outlier
df_1['outlier_Rs'] = (df_1['Rs'] < batas_bawah) | (df_1['Rs'] > batas_atas)


print(df_1['outlier_Rs'])
```

```python
# Menghitung IQR
Q1 = df_1['T'].quantile(0.25)
Q3 = df_1['T'].quantile(0.75)
IQR = Q3 - Q1

# Menentukan batas bawah dan batas atas
batas_bawah = Q1 - 1.5 * IQR
batas_atas = Q3 + 1.5 * IQR

# Menandai outlier
df_1['outlier_T'] = (df_1['T'] < batas_bawah) | (df_1['T'] > batas_atas)


print(df_1['outlier_T'])
```

```python
#Menghitung IQR
Q1 = df_1['RH'].quantile(0.25)
Q3 = df_1['RH'].quantile(0.75)
IQR = Q3 - Q1

# Menentukan batas bawah dan batas atas
batas_bawah = Q1 - 1.5 * IQR
batas_atas = Q3 + 1.5 * IQR

# Menandai outlier
df_1['outlier_RH'] = (df_1['RH'] < batas_bawah) | (df_1['RH'] > batas_atas)


print(df_1['outlier_RH'])
```
