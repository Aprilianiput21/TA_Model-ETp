```python
#Import library
import sys
import numpy as np
import pandas as pd
print(pd.__version__)
print(np.__version__)
print(sys.version)
```

```python
data1 = pd.read_csv('/content/19-07-2023.csv', skiprows = 1)
data2 = pd.read_csv('/content/AWS_30-11-23.csv', skiprows = 1)
data = pd.concat([data1, data2], ignore_index=True)
data
```

```python
file_path_output = 'AWS(Juni-November).csv'  # menyimpan data AWS Perjam
data.to_csv(file_path_output, index=False)
```

```python
#Delete useless columns
delete_columns = ['#', 'Gust Speed, m/s (LGR S/N: 21533915, SEN S/N: 21487882)', 'Wind Direction, ø (LGR S/N: 21533915, SEN S/N: 21487882)', 'Rain, mm (LGR S/N: 21533915, SEN S/N: 21506884)']
data = data.drop(columns = delete_columns)
```

```python
#Rename columns
data = data.rename(columns = { 'Date Time, GMT+08:00' : "Date Time", 'Solar Radiation, W/m² (LGR S/N: 21533915, SEN S/N: 21281112, LBL: solar)' : "Rs, W/m²", 'Wind Speed, m/s (LGR S/N: 21533915, SEN S/N: 21487882)' : "u, m/s", 'Temp, °C (LGR S/N: 21533915, SEN S/N: 21501590)' : "T, °C", 'RH, % (LGR S/N: 21533915, SEN S/N: 21501590)' : "RH, %"})
data.head()
```

```python
#Membuat kolom tanggal baru
data['Date Time'] = pd.to_datetime(data['Date Time'])
data['Date'] = data['Date Time'].dt.date
data
```

```python
#Mencari data harian (mean)
suhu = data.groupby(data['Date'])['T, °C'].mean().reset_index()
radiasi = data.groupby(data['Date'])['Rs, W/m²'].mean().reset_index()
angin = data.groupby(data['Date'])['u, m/s'].mean().reset_index()
kelembaban = data.groupby(data['Date'])['RH, %'].mean().reset_index()
suhu
```

```python
#Membuat dataframe baru
new_data = pd.concat([suhu, radiasi, angin, kelembaban],axis =1, join='outer')
new_data = new_data.round(2)
new_data
```
