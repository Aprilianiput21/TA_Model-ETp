```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('/content/PM_ET0_hourly.xlsx')
data.head()
```

```python
data = data.drop(0)
data = data.drop(1)
data.head()
```

```python
jumlah_minus = data[data['ET'] < 0]['ET'].count()

print("\nJumlah data dengan nilai 'negatif' pada kolom 'ET':", jumlah_minus)
```

```python
x = data['T']
y = data['ET']
plt.scatter(x,y)

plt.title("Korelasi Suhu dan ET")
plt.xlabel('T')
plt.ylabel('ET')
```

```python
x = data['Rs']
y = data['ET']
plt.scatter(x,y)

plt.title("Korelasi Radiasi Matahari dan ET")
plt.xlabel('Rs')
plt.ylabel('ET')
```

```python
x = data['RH']
y = data['ET']
plt.scatter(x,y)

plt.title("Korelasi Kelembaban Udara dan ET")
plt.xlabel('RH')
plt.ylabel('ET')
```

```python
x = data['U']
y = data['ET']
plt.scatter(x,y)

plt.title("Korelasi Kecepatan Angin dan ET")
plt.xlabel('U')
plt.ylabel('ET')
```
