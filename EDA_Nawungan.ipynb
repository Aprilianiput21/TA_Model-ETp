```python
#Import Library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Code menggabungkan data/file
data1 = pd.read_excel('/content/Data Nawungan_ 1.xlsx')
data2 = pd.read_excel('/content/Data Nawungan_2.xlsx')
df = pd.concat([data1, data2], ignore_index = True)
df.tail()
```

```python
#Code memisahkan kolom Date dan Time
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
df['Date'] = df['Timestamp'].dt.date
df['Time'] = df['Timestamp'].dt.time
df.head()
```

```python
#Code mengubah Time menjadi bentuk Float/Desimal
df['Time'] = df['Time'].astype(str)

# Memisahkan jam dan menit
hour_minute = df['Time'].str.split(':', expand=True)

# Mengubah jam dan menit menjadi float dan menggabungkannya
df['Time'] = hour_minute[0].astype(float) + hour_minute[1].astype(float) / 60

# Menampilkan dataframe setelah perubahan
df.head()
```

```python
#Code mendapatkan DOY
df['Date'] = pd.to_datetime(df['Date'])
df['DayOfYear'] = df['Date'].dt.dayofyear
df.head()
```

```python
df.info()
```

```python
df['u'].replace(0 and 0.0, np.nan, inplace = True)
df.tail()
```

```python
missing_data = df.isnull()
missing_data.tail()
```

```python
for column in missing_data.columns.values.tolist():
  print(column)
  print(missing_data[column].value_counts())
  print("")
```
